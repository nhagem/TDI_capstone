```python
import pandas as pd
import numpy as np
import scipy.stats
import seaborn
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#READ IN THE DATA
loan_data = pd.read_csv('C:/Users/Natasha/PycharmProjects/DataIncubator_Challenge/Data/accepted_2007_to_2018Q4.csv')
```

```python
#SIMPLIFY THE DATA
#pare down to only fully paid or charged off loans
loans = loan_data[(loan_data["loan_status"] == "Fully Paid") | (loan_data["loan_status"] == "Charged Off")]

#drop features missing >25% data
missing_data = loans.isnull().mean()
drop_list = sorted(list(missing_data[(missing_data > 0.25)].index))
loans = loans.drop(labels=drop_list, axis=1)

#drop features not available to investors, based on LCDataDictionary descriptions
keep_list = ['addr_state', 'annual_inc', 'application_type', 'dti',
             'earliest_cr_line', 'emp_length', 'emp_title', 'fico_range_high',
             'fico_range_low', 'grade', 'home_ownership', 'id',
             'initial_list_status', 'installment', 'int_rate',
             'loan_amnt', 'loan_status', 'mort_acc', 'open_acc', 'pub_rec',
             'pub_rec_bankruptcies', 'purpose', 'revol_bal', 'revol_util',
             'sub_grade', 'term', 'title', 'total_acc', 'verification_status',
             'zip_code']

drop_list2 = []
for col in loans.columns:
    if col not in keep_list:
        drop_list2.append(col)

loans = loans.drop(labels=drop_list2,axis=1)
#Remove NaN lines
loans = loans.dropna()
```

```python
#MODIFY FEATURES FOR USEFULNESS
#remove " months" from term information
loans['term'] = loans['term'].apply(lambda x: x.replace(" months", ""))
loans['term'] = loans['term'].astype(int)

#change employment length year information to integers
loans['emp_length'] = loans['emp_length'].astype(str)
loans['emp_length'] = loans['emp_length'].apply(lambda x: x.replace("< 1 year","0 years").replace("1 year","1 years").replace("10+ years","10 years").replace(" years",""))
loans['emp_length'] = loans['emp_length'].astype(int)

#simplify home ownership
loans['home_ownership'] = loans['home_ownership'].apply(lambda x: x.replace('NONE','OTHER').replace('ANY','OTHER'))

#log of annual income
loans['log_annual_inc'] = loans['annual_inc'].apply(lambda x: np.log10(x+1))

#remove outlier DTIs
loans = loans[(loans['dti'] > 0) | (loans['dti'] < 65)]

#get just year of earliest credit line
loans['earliest_cr_line'] = loans['earliest_cr_line'].apply(lambda x: int(x[-4:]))

#average fico score
loans['fico'] = 0.5*loans['fico_range_low'] + 0.5*loans['fico_range_high']

#log of credit revolving balance
loans['log_revol_bal'] = loans['revol_bal'].apply(lambda x: np.log10(x+1))

#convert loan status to 0/1 indicator
# 0 indicates fully paid, 1 indicates charged off
loans['charged_off'] = 0
loans.loc[(loans['loan_status'] == "Charged Off"),'charged_off'] = 1
```

```python
#KEEP ONLY USEFUL FEATURES
drop_list3 = ['id','grade','emp_title','annual_inc','title','zip_code',
              'fico_range_low','fico_range_high','revol_bal','loan_status']

loans = loans.drop(labels=drop_list3,axis=1)
```

```python
#EXAMINE VARIABLES FOR SIGNIFICANCE
ttest_list = ['loan_amnt','term','int_rate','installment','emp_length',
              'dti','earliest_cr_line','open_acc','pub_rec','revol_util',
              'total_acc','mort_acc','pub_rec_bankruptcies','log_annual_inc',
              'fico','log_revol_bal']

t_test_results = {}
for col in ttest_list:
    group1 = loans.where(loans.charged_off == 0).dropna()[col]
    group2 = loans.where(loans.charged_off == 1).dropna()[col]
    t_test_results[col] = scipy.stats.ttest_ind(group1,group2)

ttest_df = pd.DataFrame.from_dict(t_test_results,orient='Index')
ttest_df.columns = ['statistic','pvalue']

print(ttest_df)
```

```python
chisq_list = ['sub_grade','home_ownership','verification_status',
              'purpose','addr_state','initial_list_status','application_type']

chi_sq_results = {}
for col in chisq_list:
    con = pd.crosstab(loans[col], loans['charged_off'])
    chi2, chi_sq_results[col], dof, exp = scipy.stats.chi2_contingency(con)

chisq_df = pd.DataFrame.from_dict(chi_sq_results,orient='Index')
chisq_df.columns = ['pvalue']

print(chisq_df)
```

```python
#SPLIT INTO TESTING AND TRAINING DATA SETS

from sklearn.cross_validation import train_test_split

loans_data = loans.drop('charged_off',axis = 1)
loans_response = loans['charged_off']

#Set up dataframe with dummy variables for categorical variables
loans_dummy = pd.get_dummies(loans_data)

train_data, test_data, train_response, test_response = train_test_split(loans_dummy, loans_response, test_size=0.2)
```

```python
#CONDUCT FEATURE SELECTION USING RANDOM FOREST

from sklearn.ensemble import RandomForestRegressor as rfr

#Build forest
forest = rfr(random_state = 1, max_depth = 10)

forest.fit(train_data,train_response)
```

```python
#CHECK HOW WELL THE PREDICTION WORKED WITH ALL VARIABLES
#Using Area Under Curve due to binary response

pred = forest.predict(test_data)

from sklearn.metrics import roc_curve, auc
false_pos_rate, true_pos_rate, thresholds = roc_curve(test_response,pred)
roc_auc = auc(false_pos_rate,true_pos_rate)

print(roc_auc)
```

```python
#PLOT THE FEATURES

#NOTE: THIS IS GINI IMPORTANCE
features = train_dummy.columns
importances = forest.feature_importances_

indices = np.argsort(importances)[-9:] #top 10 features

plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
#BUILD A DATAFRAME WITH JUST THE TOP SIX

top6_train = train_data[['int_rate','term','dti','mort_acc','fico','installment']]

top6_test = test_data[['int_rate','term','dti','mort_acc','fico','installment']]

print(top6_train.head())
```

```python
#CHECK THE CORRELATION MATRIX

plt.figure(figsize=(7,6))
corr_heatmap = seaborn.heatmap(top6_train.corr(),vmin = -1, vmax = 1, cmap = 'RdBu_r')
plt.show(corr_heatmap)
```

```python
#CHECK IF ANY CORRELATIONS ARE ABOVE |0.5|

#NONE ARE

print(top6_train.corr())
```

```python
#BUILD A RANDOM FOREST USING ONLY THE TOP 6 PREDICTORS

model = rfr(random_state = 1, max_depth = 10)
model.fit(top6_train,train_response)
```

```python
#MAKE PREDICTION OF TEST DATA

prediction = model.predict(top6_test)
```

```python
#CHECK HOW WELL THE PREDICTION WORKED
#Using Area Under Curve due to binary response

from sklearn.metrics import roc_curve, auc
false_pos_rate, true_pos_rate, thresholds = roc_curve(test_response,prediction)
roc_auc = auc(false_pos_rate,true_pos_rate)

print(roc_auc)
```

```python
#PARAMATERIZE

#Vary n_estimators (number of trees in the forest)

n_estimators = [8, 16, 32, 64, 100, 200]

train_results = []
test_results = []

for n_estimator in n_estimators:
    rf = rfr(n_estimators = n_estimator, n_jobs = -1)
    rf.fit(top6_train,train_response)
    train_pred = rf.predict(top6_train)
    
    fpr, tpr, thresh = roc_curve(train_response, train_pred)
    roc_auc = auc(fpr,tpr)
    train_results.append(roc_auc)
    
    pred = rf.predict(top6_test)
    
    fpr, tpr, thresh = roc_curve(test_response, pred)
    roc_auc = auc(fpr,tpr)
    test_results.append(roc_auc)
    print(str(n_estimator) + ': completed')
    
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(n_estimators, train_results, 'b', label='Train AUC')
line2, = plt.plot(n_estimators, test_results, 'r', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('AUC score')
plt.xlabel('n_estimators')
plt.show()
```

```python
#Vary max_depth (depth of each tree in the forest)

max_depths = np.linspace(1,32,32,endpoint=True)

train_results = []
test_results = []

for max_depth in max_depths:
    rf = rfr(max_depth = max_depth, n_jobs = -1)
    rf.fit(top6_train,train_response)
    
    train_pred = rf.predict(top6_train)
    
    fpr, tpr, thresh = roc_curve(train_response,train_pred)
    roc_auc = auc(fpr,tpr)
    train_results.append(roc_auc)
    
    pred = rf.predict(top6_test)
    
    fpr, tpr, thresh = roc_curve(test_response, pred)
    roc_auc = auc(fpr,tpr)
    test_results.append(roc_auc)
    if max_depth % 5 == 0:
        print(str(max_depth) + ": completed")
    
line1, = plt.plot(max_depths, train_results, 'b', label = "Train AUC")
line2, = plt.plot(max_depths, test_results, 'r', label= "Test AUC")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('AUC score')
plt.xlabel('Tree depth')
plt.show    
```

```python
#Varying min_samples_split (minimum number of samples required
#to split an internal node). Increasing constrains the tree

min_samples_splits = [2,4,8,16,32,64,128]

train_results = []
test_results = []

for min_samples_split in min_samples_splits:
    rf = rfr(min_samples_split = min_samples_split)
    rf.fit(top6_train,train_response)
    
    train_pred = rf.predict(top6_train)
    
    fpr, tpr, thresh = roc_curve(train_response, train_pred)
    roc_auc = auc(fpr,tpr)
    train_results.append(roc_auc)
    
    pred = rf.predict(top6_test)
    
    fpr, tpr, thresh = roc_curve(test_response, pred)
    roc_auc = auc(fpr,tpr)
    test_results.append(roc_auc)
    
    print(str(min_samples_split) + ": completed")
    
line1, = plt.plot(min_samples_splits, train_results, 'b', label='Train AUC')
line2, = plt.plot(min_samples_splits, test_results, 'r', label='Test AUC')

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('AUC score')
plt.xlabel('min samples split')
plt.show()
```

```python
#Varying min_samples_leaf (minimum number of samples required to be
#at a leaf node); similar to min_samples_split

min_samples_leafs = [2,4,8,16,32,64,128]

train_results = []
test_results = []

for min_samples_leaf in min_samples_leafs:
    rf = rfr(min_samples_leaf = min_samples_leaf)
    rf.fit(top6_train,train_response)
    
    train_pred = rf.predict(top6_train)
    
    fpr,tpr,thresh = roc_curve(train_response,train_pred)
    roc_auc = auc(fpr,tpr)
    train_results.append(roc_auc)
    
    pred = rf.predict(top6_test)
    
    fpr,tpr,thresh = roc_curve(test_response,pred)
    roc_auc = auc(fpr,tpr)
    test_results.append(roc_auc)
    
    print(str(min_samples_leaf) + ": completed")
    
line1, = plt.plot(min_samples_leafs, train_results, 'b', label='Train AUC')
line2, = plt.plot(min_samples_leafs, test_results, 'r', label='Test AUC')

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('AUC score')
plt.xlabel('min sample leaf')
plt.show()
```
